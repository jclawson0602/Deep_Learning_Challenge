## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_drop_df = application_df.drop(columns=['EIN','NAME'])
application_drop_df.head()
```

```python
application_df['SPECIAL_CONSIDERATIONS'].unique()
```

```python
# Determine the number of unique values in each column.
application_drop_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_value_counts = application_drop_df['APPLICATION_TYPE'].value_counts()
application_value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Cutoff to be replaced = Less than 200 values in an application type!
application_types_to_replace = list(application_value_counts[application_value_counts < 200].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_drop_df['APPLICATION_TYPE'] = application_drop_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_drop_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_drop_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts = application_drop_df['CLASSIFICATION'].value_counts()
classification_value_counts_gt1 = classification_value_counts[classification_value_counts > 1]
classification_value_counts_gt1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Cutoff to be replaced = Less than 1000 values in a any classification!
classifications_to_replace = list(classification_value_counts[classification_value_counts < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_drop_df['CLASSIFICATION'] = application_drop_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_drop_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
preprocessed_df = pd.get_dummies(application_drop_df, dtype = int)
preprocessed_df
```

```python
# Split our preprocessed data into our features and target arrays
X = preprocessed_df.drop(columns= 'IS_SUCCESSFUL')
y = preprocessed_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model 1

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Model Changes from Original:
    # 1. Increased units in the second hidden layer from 43 to 86

nn_optimized_1 = tf.keras.models.Sequential()

# First hidden layer
nn_optimized_1.add(tf.keras.layers.Dense(units=43, activation="relu", input_dim=43))

# Second hidden layer
nn_optimized_1.add(tf.keras.layers.Dense(units=86, activation="relu"))

# Output layer
nn_optimized_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_optimized_1.summary()
```

```python
# Compile the model
nn_optimized_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
model_training = nn_optimized_1.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_optimized_1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Model Changes from Original:
    # 1. Increased units in the second hidden layer from 43 to 86
    # 2. Increased the number of epochs from 100 to 120.
    # 3. Added a third hidden layer with 86 units.

nn_optimized_3 = tf.keras.models.Sequential()

# First hidden layer
nn_optimized_3.add(tf.keras.layers.Dense(units=86, activation="relu", input_dim=43))

# Second hidden layer
nn_optimized_3.add(tf.keras.layers.Dense(units=43, activation="relu"))

# Third hidden layer
nn_optimized_3.add(tf.keras.layers.Dense(units=86, activation="relu"))

# Output layer
nn_optimized_3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_optimized_3.summary()
```

```python
# Train the model
model_training = nn_optimized_3.fit(X_train_scaled, y_train, epochs=120)
```

```python
# Evaluate the OPTIMIZED model using the test data
model_loss, model_accuracy = nn_optimized_3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
